# Let's take what we just did for one match, and do it for all of them.

```python
def make_match_treant(series):
    """Make a match Treant out of a single row of matches dataframe
    
    Parameters
    ----------
    series : pd.Series
        Pandas Series (row) giving match data
        
    Returns
    -------
    Treant
        Treant representing the match in the filesystem
    
    """
    matches = dtr.Tree('matches')
    matches.make()
    
    path = matches[str(series['matchid']) + '/']
    path
    
    match = dtr.Treant(path)
    
    match.categories['versionpatch'] = series['versionpatch']
    match.categories['map'] = series['maptype'].split('.')[1]
    match.categories['queue'] = series['queue'].split('.')[1]
    
    date = pd.DatetimeIndex([series['matchdate']])
    match.categories['day'] = int(date.day[0])
    match.categories['month'] = int(date.month[0])
    match.categories['year'] = int(date.year[0])
    match.categories['hour'] = int(date.hour[0])
    match.categories['minute'] = int(date.minute[0])
    match.categories['second'] = int(date.second[0])
    
    return match
    
    
def store_match_data(treant, series):
    """Store match data in treant as a CSV.
    
    Parameters
    ----------
    treant : Treant
        Treant corresponding to the match
    series : pd.Series
        Row giving match data
    
    """
    summoners = series['p_names'].replace('[', '').replace(']', '').replace("'", "").split(', ')
    sides = [i.split(':')[0].split('.')[1] for i in series['p_sides'].replace('[', '').replace(']', '').replace("'", "").split(', ')]
    champs = series['p_champs'].replace('[', '').replace(']', '').replace("'", "").split(', ')
    
    match_df = pd.DataFrame({'side': sides, 'champ': champs}, columns=['side', 'champ'], index=pd.Index(summoners, name='summoner'))

    match_df.to_csv(treant['match.csv'].abspath)
```

Let's make a Treant for each row:

```python
for i, row in df.iterrows():
    treant = make_match_treant(row)
    store_match_data(treant, row)
```
