## Import .csv and make into a network

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
network = pd.read_csv('WholeNetwork_Challenger_Patch6_23_Day22.csv')
Graph = nx.from_pandas_dataframe(network, 'Champion A', 'Champion B', ['Weight'])
centrality = nx.eigenvector_centrality(Graph)
plt.plot(sorted(centrality.values()))
```
